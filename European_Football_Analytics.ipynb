```python
df_1 = pd.read_sql("""

    SELECT  TRIM(LOWER(d.name)) AS Division_Name,
            TRIM(LOWER(d.country)) AS Country, 
            m.Date, 
            TRIM(LOWER(m.HomeTeam)) AS Home_Team,
            TRIM(LOWER(m.AwayTeam)) AS Away_Team,
            CAST(m.FTHG AS int) AS Home_Goals, 
            CAST(m.FTAG AS int) AS Away_Goals,
            (Home_Goals + Away_Goals) AS Total_Goals,
            m.FTR AS Result, 
            m.season AS Season

    FROM matchs AS m
    LEFT JOIN divisions AS d
    ON m.Div = d.division
    ORDER BY m.Date

""", con)
```

```python
df_2 = pd.read_sql("""

    
    
    SELECT  
            SUM(CASE WHEN Result = 'H' THEN 1 ELSE 0 END) AS 'Wins',
            COUNT(*) AS Total_Games,
            
    FROM (SELECT  
            TRIM(LOWER(d.name)) AS Division_Name,
            TRIM(LOWER(d.country)) AS Country, 
            m.Date, 
            TRIM(LOWER(m.HomeTeam)) AS Home_Team,
            TRIM(LOWER(m.AwayTeam)) AS Away_Team,
            CAST(m.FTHG AS int) AS Home_Goals, 
            CAST(m.FTAG AS int) AS Away_Goals,
            m.FTR AS Result, 
            m.season AS Season

            FROM matchs AS m
            LEFT JOIN divisions AS d
            ON m.Div = d.division
            WHERE d.country = 'France') as sub
    
    
    GROUP BY d.country

""", con)
```

```python
df_2 = pd.read_sql("""

    
    
    SELECT    

            Division_Name,
            Season,
            ROUND(SUM((CASE WHEN Result = 'H' THEN 1 ELSE 0 END) * 100.00) / COUNT(*), 2) AS 'Home_Win_Percentage',
            ROUND(SUM((CASE WHEN Result = 'A' THEN 1 ELSE 0 END) * 100.00) / COUNT(*), 2) AS 'Away_Win_Percentage'
            
            
    FROM (SELECT  
            TRIM(LOWER(d.name)) AS Division_Name,
            TRIM(LOWER(d.country)) AS Country, 
            m.Date, 
            TRIM(LOWER(m.HomeTeam)) AS Home_Team,
            TRIM(LOWER(m.AwayTeam)) AS Away_Team,
            CAST(m.FTHG AS int) AS Home_Goals, 
            CAST(m.FTAG AS int) AS Away_Goals,
            m.FTR AS Result, 
            m.season AS Season

            FROM matchs AS m
            LEFT JOIN divisions AS d
            ON m.Div = d.division) as sub
    WHERE Country = 'france'
    GROUP BY Division_Name, Season
    ORDER BY Home_Win_Percentage DESC
    LIMIT 1

""", con)
```

```python
df_2 = pd.read_sql("""

    
SELECT
        Division_Name,
        Country,
        Season,
        Home_Goals_Scored,
        Away_Goals_Scored,
        (CASE WHEN Home_Team = 'paris sg' THEN 'paris sg'
            
        
        
        
        ELSE Away_Team = 'paris sg' THEN 'paris sg' END) AS Team_Name,
        (Home_Goals_Scored + Away_Goals_Scored) AS Total_Goals_Scored
        
    
FROM(SELECT    Division_Name,
                Country,
                Season,
                SUM(CASE WHEN Home_Team = 'paris sg' THEN Home_Goals END) AS Home_Goals_Scored,
                SUM(CASE WHEN Away_Team = 'paris sg' THEN Away_Goals END) AS Away_Goals_Scored

            FROM (SELECT  
                    TRIM(LOWER(d.name)) AS Division_Name,
                    TRIM(LOWER(d.country)) AS Country, 
                    m.Date, 
                    TRIM(LOWER(m.HomeTeam)) AS Home_Team,
                    TRIM(LOWER(m.AwayTeam)) AS Away_Team,
                    CAST(m.FTHG AS int) AS Home_Goals, 
                    CAST(m.FTAG AS int) AS Away_Goals,
                    m.FTR AS Result, 
                    m.season AS Season

                    FROM matchs AS m
                    LEFT JOIN divisions AS d
                    ON m.Div = d.division) as sub
            
    WHERE Home_Team = 'paris sg' OR Away_Team = 'paris sg'
    GROUP BY Season, Country) sub_2
""", con)
```
